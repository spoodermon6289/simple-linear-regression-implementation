```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("placement.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=['package'])
y=df["package"]
```

```python
sns.scatterplot(x="cgpa", y="package", data=df,color="green")
plt.title("CGPA vs Package")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
#regression line

sns.scatterplot(x="cgpa", y="package", data=df,color="red")
plt.plot(x_test['cgpa'], y_pred, color='black', linewidth=2.2, label='Regression Line')
plt.scatter(x_test['cgpa'], y_test, color='blue', label='Actual Data')
plt.xlabel('CGPA')
plt.ylabel('Package')
plt.title('Actual vs Predicted Package')
plt.legend()
plt.show()
```

```python
m=lr.coef_#slope value
```

```python
c=lr.intercept_#y intercept
```

```python
x2=7.9
```

```python
y2=m*x2+c
y2
```

```python
#accuracy
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
